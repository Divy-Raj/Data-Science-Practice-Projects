# Number of Terrorists each Year

```python
#or
#no. of terror attacks
import seaborn as sns
plt.figure(figsize=(10,6))
fig = sns.distplot(df["Year"].values, color = 'b')
fig.set_xlabel("Year",size=15)
fig.set_ylabel("Terror attack total",size=15)
plt.title('Number of Terror attacks each year',size = 20)
plt.show()
```

# Terrorists Activites by Region in each year

```python
#Terrorist activities by region in each year
plt.subplots(figsize=(15,6))
sns.countplot('Region',data=df,palette='RdYlGn',edgecolor=sns.color_palette('dark',7),order=df['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities By Region')
plt.show()
```
